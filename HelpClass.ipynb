# 辅助类操作符

## 数学类操作符

RxJS自带的数字类操作符有四个：
* count  -- 统计数据个数
* max    -- 最大值
* min    -- 最小值
* reduce -- 规约统计

```javascript
var { Observableservable } = require('rxjs/Observable');
require('rxjs/add/observable/range');
require('rxjs/add/operator/reduce');

var source$ = Observable.range(1, 100).reduce((prev, curr) => prev+curr, 0);
source$.subscribe(console.log)
```

## 条件布尔类操作符

根据上游Observable对象的某些条件产生一个新的Observable对象

**判定函数**，参数：*被判断数据*、*序号*、*上游Observable对象*

布尔类操作符：
* `every`
* `find`
* `findIndex`
* `defaultIfEmpty`


```javascript
var { Observableservable } = require('rxjs/Observable');
require('rxjs/add/observable/of');
require('rxjs/add/observable/zip');
require('rxjs/add/observable/empty');
require('rxjs/add/operator/every');
require('rxjs/add/operator/find');
require('rxjs/add/operator/findIndex');
require('rxjs/add/operator/isEmpty');
require('rxjs/add/operator/defaultIfEmpty');

/**
* eyery 当且仅当上游传入数据全部符合判定条件时，才向下游传递true
* 遇到判定条件为false情况则立即向下游吐出flase，不在执行其他操作
* */
var source$ = Observable.of(2 ,3).every((value, index, observr$) => value >= 2);
var subscription = source$.subscribe(console.log);

/**
* find findIndex
* 找到上游Observable对象中满足判断条件的第一个数据，产生的Observable对象在吐出数据之后立刻完结
* find 吐出上游数据 
* findIndex 吐出满足判断条件的数据序列
* */
source$ = Observable.of(1, 2, 3, 4, 5, 6);
var isEven = x => x % 2 == 0;
var find$ = source$.find(isEven);
var findIndex$ = source$.findIndex(isEven);
Observable.zip(findIndex$, find$).subscribe(console.log);

/**
* isEmpty 检查上游Observable对象是不是“空”，即没有吐出任何数据就完结的Observable对象
* 对于throw（上游直接吐出error情况），isEmpty并不会处理error，而是直接丢给下游。
* 对于never，上游既不吐出数据也不完结，因此isEmpty会一直等待。
* */
source$ = Observable.empty().isEmpty();
source$.subscribe(console.log);

/**
* defaultIfEmpty
* 检查上游Observable对象是否为空，如果为空则返回默认值
* 不给defaultIfEmpty参数且上游为“空”，则向下游传递null
* */
// source$ = Observable.empty().defaultIfEmpty({});
source$ = Observable.of({name: 'dahu'}).defaultIfEmpty({});
source$.subscribe(console.log)
```
